```python
fig, ax = plt.subplots()
ax.scatter(spacer_array * 1e3, resonance_wavelength_array, color="tab:blue")
ax.set_ylabel("Resonance Wavelength", color="tab:blue")
ax.set_ylim(920, 970)
ax.tick_params(axis='y', labelcolor="tab:blue")
plt.grid()

ax2 = ax.twinx()
ax2.scatter(spacer_array * 1e3, quality_factor_array, color="tab:red")
ax2.set_ylabel("Quality Factor", color="tab:red")
ax2.set_ylim(6e3, 1.8e4)
ax2.tick_params(axis='y', labelcolor="tab:red")

ax.set_xlabel("Spacer length (nm)")
ax.set_xlim((335, 460))
plt.title("Spacer Nanobeam: radius ratio = 0.3, width = 240 nm")
plt.show()
```
